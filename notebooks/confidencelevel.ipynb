```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# For reduced chi2
chi2 = np.arange(1, 10, 0.1)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_axes([0.05, 0.15, 0.90, 0.8])
ax.set_xlabel(r"$\chi^2_{min}$")
ax.set_ylabel("n-sigma boundary")

nu_list = [10, 100, 1000]
nu_list = [1000]
for idx, nu in enumerate(nu_list):

    # 1. chi2_min * (1 + 3*(2/nu)**0.5)
    y1 = [x * (1 + 3*(2/nu)**0.5) for x in chi2]
    ax.plot(chi2, y1, ls="solid", label=(r"$\chi^2=\chi^2_{min}(1 + 3\sqrt{2/\nu})$" + r"($\nu$=" + f"{nu})"))
    
    # 2. chi2_min + 3*(nu/2)**0.5     
    y2 = [x + 3*(2/nu)**0.5 for x in chi2]
    ax.plot(chi2, y2, ls="dashed",label=(r"$\chi^2=\chi^2_{min} + 3\sqrt{2/\nu}$" + r"($\nu$=" + f"{nu})"))

    # 3. chi2_min * (1 + 1*(2/nu)**0.5)
    y1 = [x * (1 + 1*(2/nu)**0.5) for x in chi2]
    ax.plot(chi2, y1, ls="dotted",  label=(r"$\chi^2=\chi^2_{min}(1 + 1\sqrt{2/\nu})$" + r"($\nu$=" + f"{nu})"))

    # 4. chi2_min + 1*(nu/2)**0.5     
    y2 = [x + 1*(2/nu)**0.5 for x in chi2]
    ax.plot(chi2, y2, ls="dashdot", label=(r"$\chi^2=\chi^2_{min} + 1\sqrt{2/\nu}$" + r"($\nu$=" + f"{nu})"))

ax.legend()
```
